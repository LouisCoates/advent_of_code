```python
import requests
import os
from google.colab import userdata
cookie = userdata.get('AOC_SESSION_COOKIE')

def download_data(url, session_cookie, output_path):
    headers = {
        'Cookie': f'session={session_cookie}'
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        with open(output_path, 'w') as file:
            file.write(response.text)
        print(f"Data downloaded successfully and saved to {output_path}")
    else:
        print(f"Failed to download data. Status code: {response.status_code}")

if __name__ == "__main__":
    session_cookie = cookie  # Make sure to set this environment variable
    os.makedirs('data', exist_ok=True)
    if session_cookie:
          url = "https://adventofcode.com/2024/day/{}/input".format(6)
          output_path = "data/day{}_input.txt".format(6)
          download_data(url, session_cookie, output_path)
    else:
        print("Session cookie not found. Please set the AOC_SESSION_COOKIE environment variable.")

with open('data/day6_input.txt', 'r') as file:
    raw_map = file.read()
```

```python
import re

def find_guard(office):
  for i in range(len(office)):
    if '^' in list(office[i]):
      return i

def rotate_map_anti(office,num):
  for i in range(num):
    office = [list(row) for row in office]
    office = list(reversed(list(zip(*office))))
    office = [''.join(row) for row in office]
  return office

def rotate_map_cloc(office,num):
  for i in range(num):
    office = [list(row) for row in office]
    office = list(zip(*office[::-1]))
    office = [''.join(row) for row in office]
  return office
```

```python
regex = r"\^(.*?)#"
regex_end = r"\^(.*?)$"

office = raw_map.split("\n")[:-1]
guard_start_line = find_guard(office)
guard_start_posi = office[guard_start_line].index('^')



office1 = rotate_map_cloc(office,2)
turns1 = 0
while True:
  office1 = rotate_map_anti(office1,1)
  turns1 += 1
  line = find_guard(office1)
  match = re.search(regex, office1[line])
  if not match:
    match = re.search(regex_end, office1[line])
    office1[line] = office1[line][:match.start()] + 'X'*(match.end()-match.start())
    break
  else:
    office1[line] = office1[line][:match.start()] + 'X'*(match.end()-match.start()-2) + '^' + office1[line][(match.end()-1)::]

x_pos = [[pos for pos, char in enumerate(line) if char == 'X'] for line in rotate_map_cloc(office1,(turns1-2) % 4)]
x_pos[guard_start_line].remove(guard_start_posi)

print('Part 1 : ',sum([len(re.findall('X',row)) for row in office1]))

loops=0
run = 0
for row in range(len(x_pos)):
  for index in x_pos[row]:
    run += 1
    print(run , '/' , sum([len(re.findall('X',row)) for row in office1]))
    office2 = office.copy()
    office2[row] = office2[row][:index] + '#' + office2[row][(index+1)::]
    office2 = rotate_map_cloc(office2,2)
    turns2 = 0
    while True:
      office2 = rotate_map_anti(office2,1)
      turns2 += 1
      line = find_guard(office2)
      match = re.search(regex, office2[line])
      if not match:
        match = re.search(regex_end, office2[line])
        office2[line] = office2[line][:match.start()] + 'X'*(match.end()-match.start())
        break
      else:
        office2[line] = office2[line][:match.start()] + 'X'*(match.end()-match.start()-2) + '^' + office2[line][(match.end()-1)::]
      if turns2 == 1000:
        loops += 1
        break

print('Part 2 : ',loops)
```
